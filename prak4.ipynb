```python
import pandas as pd
```

```python
vaccine = pd.read_csv("vaccination_data.csv")
vaccine.head()
```

```python
covid = pd.read_csv("covid_data.csv")
covid.head()
```

```python
vaccine.info()
```

```python
vaccine['DATE_UPDATED'] = pd.to_datetime(vaccine['DATE_UPDATED'])
```

```python
vaccine.info()
```

```python
vaccine.rename(columns = {'PERSONS_VACCINATED_1PLUS_DOSE' : 'DOSE_1',
                         'PERSONS_FULLY_VACCINATED' : 'DOSE_2',
                         'PERSONS_BOOSTER_ADD_DOSE': 'BOOSTER'
                         }, inplace = True)
vaccine.info()
```

```python
df = vaccine[{'COUNTRY', 'TOTAL_VACCINATIONS','DOSE_1','DOSE_2', 'BOOSTER'}]
df.head()
```

```python
import numpy as np

df['COUNT_CATEGORY'] = np.where(
    df['TOTAL_VACCINATIONS'] > 5000000, 'Huge', np.where(
    df['TOTAL_VACCINATIONS'] < 5000000, 'Small', -1))
df.head()
```

```python
df_dropped = df.drop('COUNT_CATEGORY', axis=1)
df_dropped.head()
```

```python
covid['Date_reported'] = pd.to_datetime(covid['Date_reported'])
covid['Quarter'] = covid['Date_reported'].dt.to_period('Q')
covid_pivot = covid.pivot_table(index='Country',
                               columns = 'Quarter',
                               values = 'New_cases',
                               aggfunc = 'sum')
covid_pivot.head(10)
```

```python
covid_group1 = covid.groupby(by="Country").sum()
covid_group1.head()
```

```python
covid_group2 = covid.groupby('Country').agg({
    'New_cases': ['mean','max'],
    'New_deaths' : 'sum'})
covid_group2.head()
```

```python
covid_group = covid.groupby('Country').agg({
    'New_cases': 'sum',
    'New_deaths' : 'sum'}).reset_index()
covid_group2.head()
```

```python
covid_vaccine = covid_group.merge(vaccine,
                                 left_on = 'Country',
                                 right_on = 'COUNTRY',
                                 suffixes = (False,False),
                                 how = 'inner')
covid_vaccine
```

```python
#exercise a
death_ratio = covid[['Country','Cumulative_deaths','Cumulative_cases']]

death_ratio = covid.groupby('Country').agg({
    'Cumulative_deaths':'sum',
    'Cumulative_cases':'sum'
}).reset_index()

death_ratio['Death/case']= round((death_ratio['Cumulative_deaths']/death_ratio['Cumulative_cases'])*100,2)
death_ratio_sort = death_ratio.sort_values(by='Death/case',ascending = False)
death_ratio_sort.head()
```

```python
booster_ratio = vaccine[['COUNTRY','BOOSTER','TOTAL_VACCINATIONS']]
booster_ratio = booster_ratio.fillna(0)

booster_ratio['BOOSTER_RATIO'] = round(booster_ratio['BOOSTER']/booster_ratio['TOTAL_VACCINATIONS']*100,2)
booster_ratio_sort = booster_ratio.sort_values(by='BOOSTER_RATIO',ascending =False)
booster_ratio_sort.head()
```
